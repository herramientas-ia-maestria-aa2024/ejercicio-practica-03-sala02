```python
from faker import Faker
fake = Faker()
```

```python
lista = []
```

```python
for x in range(0,10001):
    lista.append({'nombre': fake.name(), 'apellido': fake.last_name(), 'email': fake.email(), 'address': fake.address()})
```

```python
lista
```

```python
data = pd.DataFrame(lista)
```

```python
def generar_edad(d):
    edad = random.randint(30, 40)
    return edad
```

```python
data['edad'] = data['nombre'].apply(generar_edad)
```

```python
data.head(5)
```

```python
def obtener_provincia(d):
    provincia = ['Loja', 'Pichincha', 'Guayas', 'Azuay', 'Manabí']
    prov = random.choice(provincia)
    return prov
```

```python
data['provincia'] = data['nombre'].apply(obtener_provincia)
```
